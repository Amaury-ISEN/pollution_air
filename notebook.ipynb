```python
import json
import requests
import pprint

# Nous avons généré le lien sur le site de l'API, on utilise cette URL pour requêter dessus en get
brestmet_url = "https://data.airbreizh.asso.fr/geoserver/ind_bretagne_agglo/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=ind_bretagne_agglo:ind_bretagne_agglo&outputFormat=application%2Fjson"
reponse = requests.get(brestmet_url)

# On charge le contenu de la réponse dans un JSON :
contenu = json.loads(reponse.content.decode("utf-8"))
pprint.pprint(reponse) # Réponse renvoie 200 si la requête http est OK
pprint.pprint(contenu)
```

## Création du DataFrame

Comme on a récupéré la donnée sous forme de json, le passage en DataFrame pandas est aisé avec la méthode read_json(). Il faut d'abord convertir le dictionnaire contenu en json.

```python
contenu_json = json.dumps(contenu)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(contenu["features"])
```

```python
df
```

On remarque qu'on a un DataFrame difficile à exploiter à cause des dicos nestés dans le json. On peut aplatir cela grâce à la méthode json_normalize(). Les données qui nous intéressent quant à la qualité de l'air sont surtout dans le sous dictionnaire "properties".

```python
df = pd.json_normalize(data=contenu['features'], 
                            meta=['code_zone',
                              'couleur',
                              'date_ech',
                              'lib_zone',
                              'qualif',
                              'source',
                              'type_zone',
                              'val_no2',
                              'val_o3',
                              'val_pm10',
                              'val_pm25',
                              'val_so2',
                              'valeur'])
```

```python
df
```

```python
df["properties.valeur"].value_counts()
```

On importe datetime pour changer la colonne de timestamps string en objets de type datetime.

```python
import datetime
```

```python
df["properties.date_ech"] = pd.to_datetime(df["properties.date_ech"])

# Dans datetime, %Y correspond à l'année, %U au numéro de semaine. 
df_sem = df.groupby(df["properties.date_ech"].dt.strftime('%U')).mean()

df_mois = df.groupby(df["properties.date_ech"].dt.strftime('%y/%m')).mean()

df_jour = df.groupby(df["properties.date_ech"].dt.strftime('%y/%m/%d')).mean()
```

```python
df_sem = df_sem.sort_index()
df_mois = df_mois.sort_index()
df_jour = df_jour.sort_index()

# On retire la dernière ligne de chaque dataframe, il n'y a pas assez de valeurs en 2021.
df_sem.drop(df_sem.tail(1).index,inplace=True)
df_mois.drop(df_mois.tail(1).index,inplace=True)
df_jour.drop(df_jour.tail(1).index,inplace=True)
```

## Interprétation des résultats

Il est particulièrement hardu de proposer une interprétation des résultats avec si peu de données (simples indices discrets sur une période trop courte de un an). Les autres données plus pointues disponibles sur l'API n'allaient malheureusement pas jusqu'aux périodes de confinement.

Pour être un minimum dans une démarche scientifique, il faudrait deux choses : d'une part arguer d'un effet du confinement particulier sur l'évolution de la pollution que l'on mettrait en évidence sur les courbes, d'autre part proposer un mécanisme explicatif pour prouver qu'il y a bien causalité et non pas simple corrélation entre l'évolution observée de la qualité de l'air et les confinements.

Ces deux points sont au-delà de mes capacités et je ne peux donc que faire des supputations.

Pour les points qui vont suivre, gardons à l'esprit que le premier confinement était beaucoup plus strict que le second où les écoles sont restées ouvertes et certains secteurs en activité.

**Pollution aux Oxydes d'azote (NOx)**

Pour moi, si le confinementa a pu jouer, ce serait peut-être par la diminution de la quantité de véhicules en circulation et on sait que les oxydes d'azote (NOX) sont pour 60% produits par le traffic routier. Cette forme de pollution-ci a peut-être pu diminuer un peu du fait du confinement même si c'est peu visible sur les courbes, au-moins n'y a-t-il pas eu de gros épisode NOX au cours des (et entre les) confinements.

**Pollution aux particules fines (PM10)**

On note cependant des épisodes de pollution majeurs aux particules fines en plein premier confinement et des oscillations avec pics lors des mois suivants. Ce peut être dû à des sources de pollution autres que le traffic routier comme le chauffage domestique (notamment fioul/bois) qui a dû augmenter du fait des personnes confinées à leur domicile. Le chauffage collectif avec des personnes qui travaille est réputé moins polluant qu'une démultiplication des chauffages domestiques.

**Pollution à l'ozone troposphérique (O3)**

Dans la stratosphère, l'ozone nous protège des rayonnement UV solaires nocifs. Dans la troposphère (0 à 8/15km), en revanche, il est un gaz à effet de serre et un polluant qui affecte la santé humaine à cause de ses propriétés super-oxydantes. On voit que lors du premier confinement, l'ozone n'a pas particulièrement diminué.

Difficile à expliquer dans la mesure où l'ozone est peut-être un des polluants les plus complexes dans ses sources. On considère que cette pollution provient majoritairement des NO2 produits par les échappements des moteurs, par les cheminées, les incinérateurs et incendies de forêts.

Le NO2 peut alors se photodissocier en atome d'oxygène O qui réagit avec l'O2 pour donner du O3.

Pourtant, les NO2 ne semblaient pas être très élevés lors du premier confinement. Paradoxalement, il est possible que ça soit justement ce qui a permis au O3 de remonter car normalement, la nuit, les NO2 peuvent détruire l'ozone par effet de titration et l'empêcher de trop monter.

**Phénomène global**

Les scientifiques savent aussi aujourd'hui que bien que la pollution de l'air puisse être un phénomène local et régional, c'est également un phénomène global et, parfois, la pollution peut voyager sur plusieurs jours d'un pays et même d'un continent à un autre, sur de grandes distances. Un épisode de pollution constaté en un point donné peut parfois avoir une source distante.

**Phénomène complexe**

En tout cas, la forte pollution en ozone troposphérique et en oxydes d'azote en plein premier confinement qui était pourtant particulièrement dur (rues désertes, etc.) semble suggérer que l'autmobile et le traffic routier sont loin de jouer un rôle unique dans le complexe sujet de la pollution de l'air contrairement à ce qu'on peut parfois entendre.

On peut aussi évoquer les causes naturelles de pollution atmosphérique qui sont nombreuses et peuvent parfois concurrencer ou interagir avec les émissions humaines. Par exemple des éruptions volcaniques réputées pour émettre beaucoup de dioxyde de soufre émettent également des particules fines sur de grandes distances comme cela a été démontré par des chercheurs lors de l'éruption du volcan islandais Bárðarbunga en 2014.
